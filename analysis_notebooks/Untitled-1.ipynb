```python
import numpy as np
import scipy.io as sio
import matplotlib.pyplot as plt
import scipy.signal as scisig
```

```python
path = 'E:/lorentz_force/sphere_20250327/20250328//Perm mag/2.5 V/Background/tt ('
files = [1, 2, 3, 4, 5, 6]
end = ').mat'
peak_height1 = []
for file in files:
    filename = path+str(file)+end
    data = sio.loadmat(filename)
    tt = np.transpose(data['D'])[0]
    dt = data['Tinterval'][0][0]
    for i in range(int(len(tt)/10**5)):
        PSD = scisig.welch(tt[i*10**(5):(i+1)*10**5], fs = 1/dt, nperseg = int(1/10/dt))
        peak_height1.append(PSD[1][5300])

path = 'E:/lorentz_force/sphere_20250327/20250328//Perm mag/2.5 V/Mag/tt ('
files = [1, 2, 3, 4, 5, 6]
end = ').mat'
peak_height2 = []
for file in files:
    filename = path+str(file)+end
    data = sio.loadmat(filename)
    tt = np.transpose(data['D'])[0]
    dt = data['Tinterval'][0][0]
    for i in range(int(len(tt)/10**5)):
        PSD = scisig.welch(tt[i*10**(5):(i+1)*10**5], fs = 1/dt, nperseg = int(1/10/dt))
        peak_height2.append(PSD[1][5300])
```

```python
ph_avr1 = []
for i in range(int(len(peak_height1)/10)):
    avr = np.mean(peak_height1[i*10:(i+1)*10])
    ph_avr1.append(avr)

ph_avr11 = []
for i in range(int(len(peak_height1)/100)):
    avr = np.mean(peak_height1[i*100:(i+1)*100])
    ph_avr11.append(avr)

ph_avr2 = []
for i in range(int(len(peak_height1)/10)):
    avr = np.mean(peak_height2[i*10:(i+1)*10])
    ph_avr2.append(avr)

ph_avr21 = []
for i in range(int(len(peak_height1)/100)):
    avr = np.mean(peak_height2[i*100:(i+1)*100])
    ph_avr21.append(avr)
```

```python
time1 = np.linspace(0, 300, 3000)
time2 = np.linspace(0, 300, 300)
time3 = np.linspace(0, 300, 30)
```

```python
plt.plot(time1, peak_height1, color = 'k', alpha = 0.3)
plt.plot(time2,  ph_avr1, color = 'k', alpha = 0.5)
#plt.plot(time3,  ph_avr11, color = 'k', alpha = 1)

plt.plot(time1, peak_height2, color = 'darkmagenta', alpha = 0.3)
plt.plot(time2,  ph_avr2, color = 'darkmagenta', alpha = 0.5)
#plt.plot(time3,  ph_avr21, color = 'darkmagenta', alpha = 1)

plt.yscale('log')

plt.ylabel('Peak height (V$^2$/Hz)')
plt.xlabel('Time (s)')
```

```python
path = 'E:/lorentz_force/sphere_20250321/20250325/Perm mag/2.5 V/'
files = [1, 2, 3, 4, 5, 6]
end = '.mat'
peak_height6 = []
tot_avr = []
tot_err = []
for file in files:
    filename = path+str(file)+end
    data = sio.loadmat(filename)
    tt = np.transpose(data['D'])[0]
    dt = data['Tinterval'][0][0]
    ph = []
    for i in range(int(len(tt)/10**5)):
        PSD = scisig.welch(tt[i*10**(5):(i+1)*10**5], fs = 1/dt, nperseg = int(1/10/dt))
        peak_height6.append(PSD[1][5300])
        ph.append(PSD[1][5300])
    tot_avr.append(np.mean(ph))
    tot_err.append(np.std(ph)/np.sqrt(len(ph)))
```

```python
ph_avr6 = []
for i in range(int(len(peak_height6)/10)):
    avr = np.mean(peak_height6[i*10:(i+1)*10])
    ph_avr6.append(avr)

ph_avr61 = []
for i in range(int(len(peak_height6)/100)):
    avr = np.mean(peak_height6[i*100:(i+1)*100])
    ph_avr61.append(avr)

time1 = np.linspace(0, 300, 3000)
time2 = np.linspace(0, 300, 300)
time3 = np.linspace(0, 300, 30)
```

```python
plt.plot(time1, peak_height6, 'darkmagenta', alpha = 0.3)
plt.plot(time2, ph_avr6, 'darkmagenta', alpha = 0.5)
plt.plot(time3, ph_avr61, 'darkmagenta', alpha = 1)
for i in range(50, 251, 50):
    plt.plot([i, i], [2*10**(-10), 10**(-7)], 'k')
plt.yscale('log')
plt.ylabel('Peak height (V$^2$/Hz)')
plt.xlabel('Time (s)')
plt.figure()
plt.errorbar([1, 2, 3, 4, 5, 6], tot_avr, yerr = tot_err, linestyle = 'None', marker = 'o', capsize = 8)
plt.yscale('log')
plt.ylabel('Peak height (V$^2$/Hz)')
plt.xlabel('Measurement number')
```

```python
path = 'E:/lorentz_force/sphere_20250321/20250325/Perm mag/10 V_2/'
files = [1, 2, 3, 4, 5, 6]
end = '.mat'
peak_height7 = []
nm = []
nm_err = []
m = []
m_err = []
for file in files:
    filename = path+str(file)+end
    data = sio.loadmat(filename)
    tt = np.transpose(data['D'])[0]
    dt = data['Tinterval'][0][0]
    ph = []
    for i in range(int(len(tt)/10**5)):
        PSD = scisig.welch(tt[i*10**(5):(i+1)*10**5], fs = 1/dt, nperseg = int(1/10/dt))
        peak_height7.append(PSD[1][5300])
        ph.append(PSD[1][5300])
    nm.append(np.mean(ph[:200]))
    nm_err.append(np.std(ph[:200])/np.sqrt(len(ph[:200])))
    m.append(np.mean(ph[-200:]))
    m_err.append(np.std(ph[-200:])/np.sqrt(len(ph[-200:])))
```

```python
plt.figure()
plt.errorbar([1, 2, 3, 4, 5, 6], nm, yerr = nm_err, linestyle = 'None', marker = 'o', capsize = 8, color = 'darkslategray', label = 'No magnet')
plt.errorbar([1, 2, 3, 4, 5, 6], m, yerr = m_err, linestyle = 'None', marker = 'o', capsize = 8, color = 'darkturquoise', label = 'Magnet')
plt.plot([3.5, 3.5], [5*10**(-7), 3*10**(-6)], 'k')
plt.yscale('log')
plt.ylabel('Peak height (V$^2$/Hz)')
plt.xlabel('Measurement number')
plt.legend()
```
