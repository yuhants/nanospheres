```python
# Connect to rigol

_VISA_ADDRESS_rigol = 'USB0::0x1AB1::0x0643::DG8A261500550::INSTR'
DG822 = rig.FuncGen(_VISA_ADDRESS_rigol)
```

```python
# Turn on DC

DC_volt = 0.1

DG822.DC(channel = 1, off = DC_volt)
DG822.DC(channel = 2, off = DC_volt)

DG822.turn_on(channel = 1)
DG822.turn_on(channel = 2)
```

```python
# General turn off

DG822.turn_off(channel = 1)
DG822.turn_off(channel = 2)
```

```python
# Initialise sin with offset

DC_volt = 0.1
SIN_volt = 0.002
SIN_freq = 53000

DG822.sin_wave(channel = 1, amp = SIN_volt, off = DC_volt, freq = SIN_freq, phase = 0)
DG822.sin_wave(channel = 2, amp = SIN_volt, off = DC_volt, freq = SIN_freq, phase = 180)
time.sleep(1)
DG822.turn_on(channel = 1)
time.sleep(1)
DG822.turn_on(channel = 2)
time.sleep(1)

DG822.syncronise_waveforms()

SIN_volt1 = SIN_volt
SIN_volt2 = SIN_volt
```

```python
# Ramp
#3.23/2.5
#6.455/5
#12.91/10

chan1_max_volt = 0.002
chan2_max_volt = 0.002
steps = 30

volt_step1 = (chan1_max_volt-SIN_volt1)/steps
volt_step2 = (chan2_max_volt-SIN_volt2)/steps

for i in range(steps):
    SIN_volt1 += volt_step1
    SIN_volt2 += volt_step2
    DG822.change_amplitude(channel = 1, amp = SIN_volt1)
    time.sleep(1) 
    DG822.change_amplitude(channel = 2, amp = SIN_volt2)
    time.sleep(1) 
```
